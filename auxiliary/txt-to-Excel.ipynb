```python
import re
```

```python
def find_phrase_in_file(filename, phrase):
    pattern = re.compile(re.escape(phrase), re.IGNORECASE)
    count = 0
    locations = []
    
    with open(filename, 'r') as file:
        for line_number, line in enumerate(file, start=1):
            for match in pattern.finditer(line):
                locations.append((line_number, match.start()))
                count += 1
    
    return count, locations
```

```python
filename = "Apartment-19-more-info output.txt"
phrase = 'random number 2'
count, locations = find_phrase_in_file(filename, phrase)
```

```python
print(f"'{phrase}' was found {count} times.")
if locations:
    print("Locations (line number, character position):")
    for loc in locations:
        print(f"Line {loc[0]}")
```

```python
pattern = re.compile(r"""
    (?P<time>\d{2}:\d{2}:\d{2})\s+  # Time in HH:MM:SS format
    (?P<type>sensor|room_changing)\s+  # Type: either 'sensor' or 'room_changing'
    (?:
        (?P<sensor_id>M\d{3})\s+(?P<sensor_status>ON|OFF) |  # If type is 'sensor': sensor ID and status
        (?P<from_room>\w+)\s+(?P<to_room>\w+)  # If type is 'room_changing': from room and to room
    )
""", re.VERBOSE)
```

```python
def parse_file(filename):
    with open(filename, 'r') as file:
        for line in file:
            match = pattern.match(line)
            if match:
                time = match.group('time')
                type_ = match.group('type')
                
                if type_ == 'sensor':
                    sensor_id = match.group('sensor_id')
                    sensor_status = match.group('sensor_status')
                    print(f"Time: {time}, Type: {type_}, Sensor ID: {sensor_id}, Status: {sensor_status}")
                
                elif type_ == 'room_changing':
                    from_room = match.group('from_room')
                    to_room = match.group('to_room')
                    print(f"Time: {time}, Type: {type_}, From Room: {from_room}, To Room: {to_room}")
```

```python
filename = 'Apartment-19-more-info output.txt'
parse_file(filename)
```
